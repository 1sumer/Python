```python
import numpy as np
import matplotlib.pyplot as plt

# Function to simulate the heights of students
def generate_student_heights(num_students):
    # Assume a non-normal distribution for simplicity
    return np.random.exponential(scale=10, size=num_students)

# Function to calculate the sample mean
def calculate_sample_mean(sample):
    return np.mean(sample)

# Number of students in the school
num_students_in_school = 1000

# Number of samples to take
num_samples = 50

# Size of each sample
sample_size = 30

# Simulate heights of students
student_heights = generate_student_heights(num_students_in_school)

# Simulate the distribution of sample means
sample_means = [calculate_sample_mean(np.random.choice(student_heights, size=sample_size)) for _ in range(num_samples)]

# Plot the original distribution of heights
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(student_heights, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Distribution of Heights')
plt.xlabel('Height')
plt.ylabel('Frequency')

# Plot the distribution of sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='lightcoral', edgecolor='black')
plt.title('Distribution of Sample Means')
plt.xlabel('Sample Mean Height')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
