### 1) What was the count of matches played in each season?

Number of matches played in various seasons :


```python
match_data['Season'] = pd.DatetimeIndex(match_data['date']).year
match_data.head()
```

```python
match_per_season = match_data.groupby(['Season'])['id'].count().reset_index().rename(columns={'id':'matches'})
print(match_per_season)
print(match_per_season.nsmallest(1,'matches'))
print(match_per_season.nlargest(1,'matches'))
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Season", y="matches", data=match_per_season)
plt.xlabel("Seasons",fontsize=14)
plt.ylabel("matches",fontsize=14)
plt.title("Number of matches played in various seasons ",fontsize= 20)
plt.show()
```

The number of games played per season is close to 60. But between 2011 and 2013, we are seaing slite increase in the number of matches. it is because there are now 10 teams, thanks to the addition of two new teams, the Pune Warriors and Kochi Tuskers Kerala.

### 2) How many runs were scored in each season?

Total number of runs scored across seasons

```python
season_data=match_data[['id','Season']].merge(deliveries_data, left_on = 'id', right_on = 'id', how = 'left').drop('id', axis = 1)
season_data.head()
```

```python
season = season_data.groupby(['Season'])['total_runs'].sum().reset_index()
print(season)
print(season.nsmallest(1,'total_runs'))
print(season.nlargest(1,'total_runs'))
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x="Season", y="total_runs", data=season)
plt.xlabel("Season",fontsize=14)
plt.ylabel("Total Runs",fontsize=14)
plt.title("Total Runs Across the Seasons",fontsize=18)
plt.show()
```

IN Season 2013 the highest scoring season is (22,541 runs),

#### 14) Is there any lucky venue for a particular team?

Lucky Venues for a team 

```python
def lucky(match_data,team_name):
    return match_data[match_data['winner']==team_name]['venue'].value_counts().nlargest(10)
```

```python
mi = lucky(match_data, 'Royal Challengers Bangalore')
values = mi
labels = mi.index
fig, ax = plt.subplots(figsize=(10, 6))
wedges, texts, autotexts = ax.pie(values, labels=labels, autopct='%1.1f%%', pctdistance=0.85,
                                  textprops=dict(color="w"), wedgeprops=dict(width=0.6, edgecolor='w'))
for w in wedges:
    w.set_edgecolor('w')
ax.legend(wedges, labels, title="Venues", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
ax.set_title('Wins at different Venues for RCB', fontsize=30)
plt.show()
```

#### 33) Which team has scored the most runs in the first 6 overs?

```python
team = deliveries_data['batting_team'].unique()
team_runs = []
for var in team:
    temp_df = deliveries_data[deliveries_data['batting_team']==var]
    temp_df = temp_df[temp_df['over'].isin([0,1,2,3,4,5])]
    runs = temp_df['total_runs'].sum()
    team_runs.append(runs)
team = pd.DataFrame(data=team_runs, index=team,columns=['Runs In First 6 Overs'])
#team.sort_values('Runs In First 6 Overs', ascending=False, inplace=True)
team.index.name = 'Team'

colors = ['turquoise',] * 15
colors[6] = 'crimson'

plt.figure(figsize=(10,6))
sns.barplot(x=team.index, y=team['Runs In First 6 Overs'], palette=colors)
plt.title('Total runs scored by teams in their first 6 overs', fontsize=20)
plt.xlabel('Team Name', fontsize=14)
plt.ylabel('Total runs', fontsize=14)
plt.xticks(rotation=90)
plt.show()
```
