### **Creating Features**

We will manipulate the variables related to age in our dataset. Upon examining the sample data, the 'Age' column represents the age of a person.

However, it might be challenging to understand how age relates to survival. 

To address this, we will introduce a new column, 'Age_group', to categorize the ages of individuals.

```python
def categorize(age):
    if age < 17.0:
        return 'Young'
    elif age<34.0:
        return 'Adult'
    elif age< 54.0:
        return 'Middle age'
    else:
        return 'Seniors'  
```

```python
df['Age_group'] = df['age'].apply(categorize)
```

```python
plt.figure(figsize=(12,6))

ax = plt.subplot(1,2,1)
ax = sns.countplot(x='survived', data=df)
ax.set_xticklabels(['Unsurvived', 'Survived'])
ax.bar_label(ax.containers[0])
plt.title("Survived", fontsize=20, fontweight= 'bold')

ax = plt.subplot(1,2,2)
ax=df['survived'].value_counts().plot.pie(labels = ['Unsurvived','Survived'],explode=[0, 0.2],autopct='%1.f%%',shadow=True)
ax.set_title(label = "Survived", fontsize = 20,color='Red',font='Lucida Calligraphy', fontweight= 'bold')

plt.show()
```
