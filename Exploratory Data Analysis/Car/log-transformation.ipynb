## **Transformation**

Logarithmic transformation is a mathematical operation applied to a set of data points to change their scale by taking the logarithm of each point. The most commonly used base for logarithmic transformations is the natural logarithm, which uses the base "e" (approximately 2.71828).

The formula for the natural logarithm of a number "x" is:

<code>log10 (x)<br>
loge (x)</code>

This operation helps in compressing large ranges of values into a smaller range, making it easier to visualize and analyze data, especially when dealing with skewed distributions or data with a wide range of values. Logarithmic transformations are particularly useful in cases where the data is highly skewed

#### **Here's how it works:**

**Compression of large values:** Logarithmic transformation compresses large values more than small ones. This means that extreme values are reduced in importance relative to smaller values. For example, the difference between 100 and 1,000 is much smaller after a logarithmic transformation compared to the original scale.

#### **When to use log transformation:**

**Skewed and Outliers data:** When the data is heavily skewed, such as in the case of income distribution or population sizes.


Example:

Consider a dataset containing the population sizes of different cities. Some cities might have very large populations (e.g., millions of people), while others might have relatively small populations (e.g., thousands of people). This dataset could be highly skewed towards the larger cities.

By applying a logarithmic transformation to the population sizes, we can compress the range of values. For example, if we take the natural logarithm of each population size, a city with a population of 1,000,000 might transform to approximately 13.82, while a city with a population of 10,000 might transform to approximately 9.21. This compression makes it easier to visualize and analyze the data, particularly when comparing cities of vastly different sizes.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset representing two columns: population sizes and economic output
data = np.array([[1000, 50000], 
                 [10000, 100000], 
                 [100000, 500000], 
                 [1000000, 1000000], 
                 [10000000, 5000000]])

# Applying base 10 logarithm transformation to both columns
log10_data = np.log10(data)

print(data)
print(log10_data)
## Reverting the values back to their original scale
#original_data = np.power(10, log10_data)

# Plotting distribution plots
plt.figure(figsize=(10, 5))

# Original data distribution plots
plt.subplot(1, 2, 1)
sns.distplot(data[:, 0], label='Population Sizes')
sns.distplot(data[:, 1], label='Economic Output')
plt.title('Original Data Distribution')
plt.xlabel('Values')
plt.legend()

# Log-transformed data distribution plots
plt.subplot(1, 2, 2)
sns.distplot(log10_data[:, 0], label='Population Sizes')
sns.distplot(log10_data[:, 1], label='Economic Output')
plt.title('Log-transformed Data Distribution (Base 10)')
plt.xlabel('Log-transformed Values (Base 10)')
plt.legend()

plt.tight_layout()
plt.show()
```
