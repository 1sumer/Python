### **Flipkart Autoscraping for Apples Phone**

```python
from autoscraper import AutoScraper
```

```python
amazon_url = "https://www.flipkart.com/search?q=iphone"

wanted_list = ["Apple iPhone 14 Plus (Starlight, 128 GB)","₹55,999","72,870 Ratings & 4,096 Reviews"]
```

```python
scraper = AutoScraper()
result = scraper.build(amazon_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(amazon_url,grouped=True)
```

```python
rule_aliases = {
    'rule_od2b': 'Title',  # Rule ID for titles
    'rule_7fmm': 'Price'   # Rule ID for prices
}
scraper.set_rule_aliases(rule_aliases)
```

```python
result = scraper.get_result_similar("https://www.flipkart.com/search?q=mi+phone+under+15000", group_by_alias=True)
```

```python
title = result['Title']
price = result['Price']
```

```python
sameer = (title,price)
sameer
```

```python
import pandas as pd
data1={'Title':title, 'FPrice':price}
df = pd.DataFrame(data1)
print(df)
```

```python
df.to_csv('Mi_4.csv')
```
