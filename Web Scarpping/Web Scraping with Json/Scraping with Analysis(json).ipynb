```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
URL = "https://pythonjobs.github.io/"
page = requests.get(URL)
page
```

```python
soup = BeautifulSoup(page.content, "html.parser")
print(soup)
```

```python
results = soup.find(id="container")
print(results)
```

```python
job_elements = results.find_all("div", {"class":"job"})
print(job_elements)
```

```python
title = []
company = []
location = []
date = []
job_type = []
```

```python
for job_element in job_elements:
    name = job_element.find("h1").text.strip()
    address = job_element.find('span', class_='info').text.strip()
    time = job_element.find_all('span', class_='info')[1].text.strip()
    job = job_element.find_all('span', class_='info')[2].text.strip()
    department = job_element.find_all('span', class_='info')[3].text.strip()
    
    title.append(name)
    company.append(department)
    location.append(address)
    date.append(time)
    job_type.append(job)
```

```python
print(title[0])
print(company[0])
print(location[0])
print(date[0])
print(job_type[0])
```

```python
dataframe = {'Job Title':title, 'Company':company, 'Location':location, 'Date':date, 'Job Type': job_type}

# Saving scraped data to a JSON file
with open('products.json', 'w') as file:
    json.dump(dataframe, file, indent=4)
```
