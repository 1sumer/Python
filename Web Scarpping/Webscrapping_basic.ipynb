```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.bbc.com/news'
response = requests.get(url)
response
```

```python
# print content of request
print(response.content)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
# Getting the title tag
print(soup.title)
```

```python
headlines = soup.find('body').find_all('h2')
paragraph = soup.find('body').find_all('p') 
```

```python
for x, y in zip(headlines,paragraph): 
    print("\033[1m" + x.text.strip() + "\033[0m")
    print(f'{y.text.strip()}\n')
```

#### **Links**

```python
links = soup.find('body').find_all('a', href=True)
for x in links:
    if x['href'].startswith('/news/'):
        print(f'https://www.bbc.com{x['href']}')
```

#### **Most Watch**

```python
watch = results.find_all("div", {"class":"sc-4fedabc7-1 iqyyKX"})
print(watch)
```

```python
for most_watch in watch:
    headings = most_watch.find_all("h2", {"class":"sc-4fedabc7-3 gJvjEE"})
    for heading in headings:
        heading_text = heading.text.strip()
        print(heading_text)
```
