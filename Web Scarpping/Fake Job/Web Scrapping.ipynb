### Web Scrapping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.bbc.com/news'
response = requests.get(url) 
  
soup = BeautifulSoup(response.text, 'html.parser') 
headlines = soup.find('body').find_all('h2') 
for x in headlines: 
    print(x.text.strip()) 
```

## **Web Scraping the Website for Fake Jobs**

```python
URL = "https://realpython.github.io/fake-jobs/"
page = requests.get(URL)
page
```

```python
soup = BeautifulSoup(page.content, "html.parser")
print(soup)
```

```python
results = soup.find(id="ResultsContainer")
print(results)
```

```python
job_elements = results.find_all("div", {"class":"card-content"})
print(job_elements)
```

```python
title = []
company = []
location = []
time = []
```

```python
for job_element in job_elements:
    title_element = job_element.find("h2", class_="title is-5").text.strip()
    company_element = job_element.find("h3", class_="subtitle is-6 company").text.strip()
    location_element = job_element.find("p", class_="location").text.strip()
    date = job_element.find("p", class_="is-small has-text-grey").text.strip()
    title.append(title_element)
    company.append(company_element)
    location.append(location_element)
    time.append(date)
```

```python
print(title[0])
print(company[0])
print(location[0])
print(time[0])
```

```python
dataframe = {'Job Title':title, 'Company':company, 'Location':location, 'Date':time}
final_dataframe = pd.DataFrame(dataframe)
final_dataframe 
```

```python
final_dataframe.to_excel('Products.xlsx')
```
