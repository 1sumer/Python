Linear regression involves finding the best-fit line for a set of data points. The equation for a simple linear regression is typically written as:

<code>y=mx+b</code>

where 
* y is the dependent variable, 
* x is the independent variable, 
* m is the slope of the line, and 
* b is the y-intercept.

* weeks contains the dependent variable
* sales contains the independent variable

#### Type 1

```python
# Sample data
weeks = [1, 2, 3, 4, 5]
sales = [10, 14, 18, 24, 30]
```

calculate the mean (average) of the weeks and sales

```python
# Calculate the mean of weeks and sales
mean_weeks = sum(weeks) / len(weeks)
mean_sales = sum(sales) / len(sales)
```

- calculate the slope (m) and y-intercept (b)
- The numerator represents the sum of the product of the differences between each data point and the mean for both weeks and sales.
- denominator represents the sum of the squared differences between each week value and the mean week.

```python
# Calculate the slope (m) and y-intercept (b)
numerator = sum((xi - mean_weeks) * (yi - mean_sales) for xi, yi in zip(weeks, sales))
denominator = sum((xi - mean_weeks)**2 for xi in weeks)
```

<code>The formula for the slope (m) is given by:<br>
m=n(∑x*y) −(∑x) (∑y) ​/n(∑x2) −(∑x)2</code>

```python
slope = numerator / denominator
intercept = mean_sales - slope * mean_weeks
```

```python
# Print the results
print(f"Slope (m): {slope}")
print(f"Y-Intercept (b): {intercept}")
```

```python
# Predict sales for a new week
new_week = 6
predicted_sales = slope * new_week + intercept
print(f"Predicted Sales for Week {new_week}: {predicted_sales}")
```

```python
import matplotlib.pyplot as plt
# Plot the original data and the linear regression line
plt.scatter(weeks, sales, label='Original Data')
plt.plot(weeks, [slope * xi + intercept for xi in weeks], label='Linear Regression', color='red')
plt.xlabel('Weeks')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]

# Calculate the mean of x and y
mean_x = sum(x) / len(x)
mean_y = sum(y) / len(y)

# Calculate the slope (m) and y-intercept (b)
numerator = sum((xi - mean_x) * (yi - mean_y) for xi, yi in zip(x, y))
denominator = sum((xi - mean_x)**2 for xi in x)

slope = numerator / denominator
intercept = mean_y - slope * mean_x

# Print the results
print(f"Slope (m): {slope}")
print(f"Y-Intercept (b): {intercept}")
```

#### Type 2

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 14, 18, 24, 30])

# Fit a linear regression model
# Create a linear regression function
linear_function = np.poly1d(np.polyfit(x, y, 1))

# Generate x values for the fitted line
x_fit = np.linspace(min(x), max(x), 100)

# Calculate corresponding y values using the linear function
y_fit = linear_function(x_fit)

# Plot the original data and the fitted line
plt.scatter(x, y, label='Original Data')
plt.plot(x_fit, y_fit, label='Linear Regression', color='red')
plt.xlabel('Weeks')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

#### Type 3

```python
import statsmodels.api as sm

X = x
y = y

# Add a constant term to the independent variable (income)
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the summary
print(model.summary())

# Predicted values based on the model
y_pred = model.predict(X)
```
