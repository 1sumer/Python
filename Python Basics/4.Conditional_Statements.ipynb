## **Python Conditional Statements:**
Conditional statements in Python are used to execute different blocks of code based on whether a specified condition evaluates to True or False. They are fundamental to programming as they enable the implementation of decision-making logic within programs. Python provides several conditional statements to facilitate this, namely if, else, elif, and nested conditionals.

### **1.if Statement:**
The if statement is the most basic conditional statement in Python. It allows you to execute a block of code only if a specified condition is true.

**Syntax**

<pre><code>if condition:
    # block of code to be executed if condition is true </code></pre>

```python
x = 10
if x > 5:
    print("x is greater than 5")
```

```python
x = 10
if x < 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")
```

```python
a = False #try assigning False to a
if a:
    print('Hello') #try removing the indentation
else:
    print('Bye')
```

### **2.else Statement:**
The else statement provides an alternative action for when the condition in the if statement is not met.
- If the condition is true: Python runs the code inside the if block.
- If the condition is false: Python skips the code inside the if block and moves to the else block, running the code inside it.


**Syntax :**

<pre><code>if condition:
    # block of code to be executed if condition is true
else:
    # block of code to be executed if condition is false</code></pre>

#### FLOW CHART:


Conditional expression should only be of the form of Boolean. The output of the conditional expression can only either be TRUE or FALSE.

```python
x = 5
if x % 2 == 0:
    print("x is even")
else:
    print("x is odd")
```

### **3.elif Statement:**
The elif statement, short for "else if", allows you to check multiple conditions sequentially. It is used when you have more than two possible outcomes.

<pre><code>if condition1:
    # block of code to be executed if condition1 is true
elif condition2:
    # block of code to be executed if condition1 is false and condition2 is true
else:
    # block of code to be executed if all conditions are false</code></pre>


```python
a=5
b=5
if a == b:
    print("How did that happen?")
elif a > b:
    print("Yikes")
else:
    print("All is well with the world")
```

```python
d = [1,2,3,"true"]
if d[-3]>2:
    print("hay i am here")
elif d[-3]>=2:
    print("dont ask")
else:
    print("i am outside")
```

### **4.Nested if Statements:**
Python allows you to nest conditional statements within one another, creating more complex decision-making logic.

Nested if statements in Python are like opening a series of boxes inside another box. Each box (or if statement) contains its own instructions, and you can have one box inside another.

<pre><code>if condition1:
    # block of code
    if condition2:
        # block of code
    else:
        # block of code
else:
    # block of code</code></pre>


```python
x = 10
if x > 5:
    print("x is greater than 5")
    if x == 10:
        print("x is also equal to 10")
else:
    print("x is less than or equal to 5")
```

```python
x = -4
if x>0:
    if x%2==0:
        print(x,"is even number")
    else:
        print(x,"is odd number")
elif x<0:
    print(x,"is negative number")
else:
    print(x,"is zero")
```

```python
a = float(input("Enter the length of side a: "))
b = float(input("Enter the length of side b: "))
c = float(input("Enter the length of side c: "))

if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("Equilateral triangle")
    elif a == b or b == c or a == c:
        print("Isosceles triangle")
    else:
        print("Scalene triangle")
else:
    print("Not a valid triangle")
```

```python
print("Simple Calculator")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

choice = input("Enter choice (1/2/3/4): ")

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

if choice == '1':
    result = num1 + num2
    print("Result:", result)
elif choice == '2':
    result = num1 - num2
    print("Result:", result)
elif choice == '3':
    result = num1 * num2
    print("Result:", result)
elif choice == '4':
    if num2 != 0:
        result = num1 / num2
        print("Result:", result)
    else:
        print("Error: Division by zero")
else:
    print("Invalid choice")
```

**Notes:**

- **Indentation:** In Python, indentation is used to define blocks of code. All statements within the same block must be indented consistently.
- **Logical Operators:** You can use logical operators (and, or, not) to combine multiple conditions within a conditional statement.
- **Multiple Conditions:** You can have as many elif clauses as needed to handle various conditions.
- **Execution Flow:** Python evaluates the conditions sequentially, executing the block of code associated with the first true condition encountered.
