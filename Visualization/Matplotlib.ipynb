## **MatPlotLib**

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK. There is also a procedural "pylab" interface based on a state machine (like OpenGL), designed to closely resemble that of MATLAB, though its use is discouraged. SciPy makes use of Matplotlib.

Matplotlib was originally written by John D. Hunter.

Matplotlib supports various types of 2 dimensional and 3 dimensional plots. The support for two dimensional plots is robust. The support for three dimensional plots was added later and while it is good, it is not as robust as 2 dimensional plots.

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
#pip install matplotlib, numpy, scipy
```

```python
x = np.arange(-3, 3, 0.1)
print(x)
print("-----------------------------------------------------------------------")
y = norm.pdf(x)
print(y)
```

```python
plt.scatter(x,y) #Scatter Plot
plt.title("Normal Distribution")
plt.show() # Displaying Plot
```

```python
plt.plot(x,y) #Line Plot
plt.show()
```

### **Mutiple Plots on One Graph**

```python
plt.scatter(x, y) # Scaytter plot
plt.plot(x, norm.pdf(x, 1, 0.5)) # line Plot
plt.show() # Show the plot
```

### **Labeling Axes and Adding a Legend**

```python
plt.plot(x, x**2, color='blue', linestyle=':')
plt.plot(x, norm.pdf(x), color='yellow', marker='*')
plt.plot(x, norm.pdf(x,1.0,0.5), 'g+')

plt.xlim([-4, 4]) # adjusting x axis value range
plt.ylim([0, 2.0]) # adjusting y axis value range

plt.grid(axis='both', color='k', linestyle=':',linewidth=1)

plt.xlabel('Event1')
plt.ylabel('Event2')

plt.title('line Graph',fontsize=15, color='b', loc='center', fontweight='bold')

plt.legend(['Fun A', 'Fun B', 'Fun C'], loc='best') #, 'best', 'upper left', 'upper right', 'lower left', 'lower right', 1, 2, 4

plt.show()
```

### **Pie Chart**

```python
values = [12, 30, 4, 32, 14]

colors = ['r', 'g', 'b', 'c', 'm']

explode = [0.2, 0, 0.5, 0, 0]

labels = ['India', 'United States', 'Russia', 'China', 'Europe']

plt.pie(values, radius=2, colors= colors, labels=labels, explode=explode, autopct='%1.1f%%',frame=False)
plt.axis('equal')

plt.legend(title='country', loc='best') #loc=4
plt.title('Student Locations')
plt.tight_layout()

plt.show()
```

### **Bar Chart**

```python
import seaborn as sns
x = [1, 5, 6, 7, 8]

y = [12, 55, 4, 32, 14]

#colors = ['r', 'g', 'b', 'c', 'm']

# Plotting the bar chart
bars = plt.bar(x, y, color=sns.color_palette('Set2'), edgecolor='k', align='center')
plt.plot(x,y, marker='o', color='r')

# Adding labels
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Adding a title
plt.title('Bar Chart Example')

# Adding a legend
plt.legend(bars, ['Bar 1', 'Bar 2', 'Bar 3', 'Bar 4', 'Bar 5'])

# Add bar labels with vertical rotation and 2 decimal places
plt.bar_label(bars, label_type='center', rotation=90, fmt='%.2f') #label_type = edge, fmt is displaying 2 decimal values

# Adding a grid
plt.grid(axis='both', color='k',linestyle='--', linewidth=0.5)

plt.savefig(r"D:\Python\Visualization\Plot.png", format='png')

```

```python
x = [1, 5, 8, 6, 7]

y = [12, 55, 4, 32, 14]

# Plotting the bar chart
bars = plt.barh(x, y, color=sns.color_palette('Set2'), edgecolor='k', align='center')

# Adding labels
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Adding a title
plt.title('Bar Chart Example')

# Adding a legend
plt.legend(bars, ['Bar 1', 'Bar 2', 'Bar 3', 'Bar 4', 'Bar 5'])
plt.bar_label(bars, label_type='center') #edge

# Set the y-axis to plain numerical format (no scientific notation)
plt.ticklabel_format(style='plain', axis='y')

# Adding a grid
plt.grid(axis='both', color='k',linestyle='--', linewidth=0.5)

plt.show()
```

### **Box & Whisker Plot**

Useful for visualizing the spread & skew of data.

The red line represents the median of the data, and the box represents the bounds of the 1st and 3rd quartiles.

So, half of the data exists within the box.

The dotted-line "whiskers" indicate the range of the data - except for outliers, which are plotted outside the whiskers. Outliers are 1.5X or more the interquartile range.

This example below creates uniformly distributed random numbers between -40 and 60, plus a few outliers above 100 and below -100:

<pre><code>     Q1-1.5IQR   Q1   median  Q3   Q3+1.5IQR
                  |-----:-----|
  o      |--------|     :     |--------|    o  o
                  |-----:-----|
flier             <----------->            fliers
                       IQR</code></pre>

```python
import pandas as pd
uniformSkewed = np.random.rand(100) * 100 - 40
high_outliers = np.random.rand(10) * 50 + 100
low_outliers = np.random.rand(10) * -50 - 100
data = np.concatenate((uniformSkewed, high_outliers, low_outliers))
df = pd.DataFrame({'data': data}, columns=['data'])
print(df)
df.describe()
```

```python
plt.boxplot(df['data'],labels=df.columns, vert=True, showfliers=True, meanline=True)
plt.show()
```

### **Subplots**

```python
x=np.arange(-3,3,0.1)
data = np.random.randn(1000)

plt.figure(figsize=(8,6)) 

plt.subplot(2, 2, 1)  # This is to create a 2x2 plot 1st position
plt.plot(x,x**2, color='blue', linestyle='--')
plt.xlabel('Event1')
plt.ylabel('Event2')
plt.title('Fun A')

plt.subplot(2, 2, 2)
plt.scatter(x, norm.pdf(x), color='yellow', marker='*')
plt.xlabel('Event1')
plt.ylabel('Event2')
plt.title('Fun B')

plt.subplot(2, 2, 3)
plt.hist(data, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Event1')
plt.ylabel('Event2')
plt.title('Fun C')

plt.tight_layout()  # This is to make the plot look nice
plt.show()
```
