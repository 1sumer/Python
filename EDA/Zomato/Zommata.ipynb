* Zomato has Launched in 2010, its technology platform connects customers, restaurant partners and delivery partners, serving their multiple needs.

```python
df.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
columns_lst=list(df.columns)
for i in columns_lst:
    print('Number of unique values in',i, 'is',df[i].nunique())
```

```python
plt.figure(figsize=(10,20))
for i in range(len(columns_lst)):
    plt.subplot(5,3,i+1)
    plt.title(columns_lst[i])
    plt.xticks(rotation=45)
    plt.hist(df[columns_lst[i]])
    
plt.tight_layout()
```

```python
RATING = df['rate'].value_counts().head()
RATING
```

```python
RATING.plot(kind='pie',autopct='%1.f%%')
plt.title('rating')
plt.show()
```

```python
PLT = df[df['rate']==3.9].value_counts()[:20]
sns.set_style(style='whitegrid')
plt.figure(figsize=(5,5))
sns.distplot(PLT, label='Rate', color='tomato', kde=True)
plt.title('RATING')
plt.legend(loc=1)
plt.grid()
plt.show()
```

```python
df_su = pd.DataFrame(df.groupby('votes')['votes'].count())
df_su.head()
```

```python
df_hu = df.groupby('votes')['votes'].count()[:50]
sns.countplot(df_hu,label='Votes', color='tomato')
plt.title('Votes')
plt.legend(loc=1)
plt.grid()
```

```python
sns.set_style(style='whitegrid')
plt.figure(figsize=(5,5))
sns.histplot(df['votes'], label='Votes', color='tomato', kde=True)
plt.title('Votes')
plt.legend(loc=1)
plt.grid()
```
