```python
# categorical column
# imput the values using mode
cat_mode = car_data2['Category'].mode()[0]
cat_mode
```

```python
print(car_data2['No_of_Doors'].mode())
```

```python
car_data2['No_of_Doors'] = car_data2['No_of_Doors'].fillna(car_data2['No_of_Doors'].mode()[0])
```

```python
car_data2['Fuel_Type'].mode()
```

```python
car_data2['Fuel_Type'] = car_data2['Fuel_Type'].fillna(car_data2['Fuel_Type'].mode()[0])
```

### **EDA Univariate Analysis**

Analyzing/visualizing the dataset by taking one variable at a time:

Data visualization is essential; we must decide what charts to plot to better understand the data. In this article, we visualize our data using Matplotlib and Seaborn libraries.

Matplotlib is a Python 2D plotting library used to draw basic charts we use Matplotlib.

Seaborn is also a python library built on top of Matplotlib that uses short lines of code to create and style statistical plots from Pandas and Numpy

Univariate analysis can be done for both Categorical and Numerical variables.

Categorical variables can be visualized using a Count plot, Bar Chart, Pie Plot, etc.

Numerical Variables can be visualized using Histogram, Box Plot, Density Plot, etc.

In our example, we have done a Univariate analysis using Histogram and  Box Plot for continuous Variables.

In the below fig, a histogram and box plot is used to show the pattern of the variables, as some variables have skewness and outliers.

```python
for col in num_cols:
    print(col)
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    CarData[col].hist(grid=False)
    plt.ylabel('count')
    plt.show()
```

```python
fig, axes = plt.subplots(4, 2, figsize = (18, 24))

fig.suptitle('Bar plot for all categorical variables in the dataset')

sns.countplot(ax = axes[0, 0], x = 'Make', data = CarData, color = 'blue', order = CarData['Make'].value_counts().head(10).index)

sns.countplot(ax = axes[0, 1], x = 'Model', data = CarData, color = 'blue', order = CarData['Model'].value_counts().head(10).index)

sns.countplot(ax = axes[1, 0], x = 'Fuel_Type', data = CarData, color = 'blue',  order = CarData['Fuel_Type'].value_counts().index)

sns.countplot(ax = axes[1, 1], x = 'Transmission', data = CarData, color = 'blue',  order = CarData['Transmission'].value_counts().index)

sns.countplot(ax = axes[2, 0], x = 'Driven_Wheels', data = CarData, color = 'blue', order = CarData['Driven_Wheels'].value_counts().index)

sns.countplot(ax = axes[2, 1], x = 'Category', data = CarData, color = 'blue', order = CarData['Category'].value_counts().head(10).index)

sns.countplot(ax = axes[3, 0], x = 'Size', data = CarData, color = 'blue', order = CarData['Size'].value_counts().index)

sns.countplot(ax = axes[3, 1], x = 'Style', data = CarData, color = 'blue', order = CarData['Style'].value_counts().head(10).index)

for ax in axes.flatten():
    for p in ax.patches:
        ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

axes[0][0].tick_params(labelrotation=45)
axes[0][1].tick_params(labelrotation=45)
axes[1][0].tick_params(labelrotation=75)
axes[1][1].tick_params(labelrotation=45)
axes[2][0].tick_params(labelrotation=30)
axes[2][1].tick_params(labelrotation=45)
axes[3][1].tick_params(labelrotation=90)
```

```python
fig, axarr = plt.subplots(6, 2, figsize=(18, 24))

# Manufacture Vs Price
CarData.groupby('Make')['MSRP'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[0][0], fontsize=12, color='skyblue')
axarr[0][0].set_title("Manufacture Vs Price", fontsize=14)

# Transmission Vs Price
CarData.groupby('Transmission')['MSRP'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][1], fontsize=12, color='lightcoral')
axarr[0][1].set_title("Transmission Vs Price", fontsize=10)

# Fuel_Type Vs Price
CarData.groupby('Fuel_Type')['MSRP'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][0], fontsize=12, color='lightgreen')
axarr[1][0].set_title("Fuel_Type Vs Price", fontsize=14)

# No_of_Doors Vs Price
CarData.groupby('No_of_Doors')['MSRP'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][1], fontsize=12, color='gold')
axarr[1][1].set_title("No_of_Doors Vs Price", fontsize=14)

# Brand Vs Price
CarData.groupby('Style')['MSRP'].mean().sort_values(ascending=False).plot.bar(ax=axarr[2][0], fontsize=12, color='lightblue')
axarr[2][0].set_title("Brand Vs Price", fontsize=14)

# Model Vs Price
CarData.groupby('Model')['MSRP'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[2][1], fontsize=12, color='lightcoral')
axarr[2][1].set_title("Model Vs Price", fontsize=14)

# Size Vs Price
CarData.groupby('Size')['MSRP'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[3][0], fontsize=12, color='lightgreen')
axarr[3][0].set_title("Size Vs Price", fontsize=14)

# Car_Age Vs Price
CarData.groupby('Car_Age')['MSRP'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][1], fontsize=12, color='gold')
axarr[3][1].set_title("Car_Age Vs Price", fontsize=14)

# Popularity Vs Price
CarData.groupby('Popularity')['MSRP'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[4][0], fontsize=12, color='gold')
axarr[4][0].set_title("Popularity Vs Price", fontsize=14)

# EngineHP Vs Price
CarData.groupby('EngineHP')['MSRP'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[4][1], fontsize=12, color='gold')
axarr[4][1].set_title("EngineHP Vs Price", fontsize=14)

# Highway_MPG Vs Price
CarData.groupby('Highway_MPG')['MSRP'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[5][0], fontsize=12, color='gold')
axarr[5][0].set_title("Highway_MPG Vs Price", fontsize=14)

# City_mpg Vs Pric
CarData.groupby('City_mpg')['MSRP'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[5][1], fontsize=12, color='gold')
axarr[5][1].set_title("City_mpg Vs Price", fontsize=14)


plt.subplots_adjust(hspace=1.0)
plt.subplots_adjust(wspace=.5)
sns.despine()
```
