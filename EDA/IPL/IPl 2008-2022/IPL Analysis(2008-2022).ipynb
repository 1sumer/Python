### 1) What is the count of matches played in each season?

Number of matches played in various seasons :


```python
match_data['Season'] = pd.DatetimeIndex(match_data['Date']).year
match_data.head()
```

```python
match_data.groupby(['Season'])['ID'].count()
match_data.groupby(['Season'])['ID'].count().reset_index()
match_data.groupby(['Season'])['ID'].count().reset_index().rename(columns={'ID':'Matches'})
```

```python
match_per_season = match_data.groupby(['Season'])['ID'].count().reset_index().rename(columns={'ID':'matches'})
print(match_per_season)
print(match_per_season.nsmallest(1,'matches'))
print(match_per_season.nlargest(1,'matches'))
```

```python
plt.figure(figsize=(9, 6))
sns.barplot(x="Season", y="matches", data=match_per_season, palette="rocket_r")
plt.xlabel("Seasons",fontsize=14)
plt.ylabel("matches",fontsize=14)
plt.title("Number of matches played in various seasons ",fontsize= 20, fontweight="bold")
plt.show()
```

The number of games played per season is close to 60. But between 2011, 2013 & 2022, we are seaing slite increase in the number of matches. it is because there are now 10 teams, thanks to the addition of two new teams, the Pune Warriors and Kochi Tuskers Kerala.

### 2) How many runs were scored in each season?

Total number of runs scored across seasons

```python
season_data = match_data[['ID','Season']].merge(deliveries_data, left_on = 'ID', right_on = 'ID', how = 'left').drop('ID', axis = 1)
season_data.head()
```

```python
season = season_data.groupby(['Season'])['total_run'].sum().reset_index()
print(season)
print(season.nsmallest(1,'total_run'))
print(season.nlargest(1,'total_run'))
```

### 3) What were the total number of matches played and the total runs scored in each season?

```python
runs_per_season=pd.concat([match_per_season,season.loc[:,'total_run']],axis=1)
print(runs_per_season)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(runs_per_season['Season'], runs_per_season['total_run'], marker='o', linestyle='-.', color='green')
plt.xticks([2009,2013, 2022])
plt.title('Total Runs Scored per Season')
plt.xlabel('Season')
plt.ylabel('Total Runs')
#plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
#plt.tight_layout()
plt.show()
```

- IN Season 2022 the highest scoring season is (24,395 runs),
- IN Season 2009 the lowest scoring season is (16,353 runs).
