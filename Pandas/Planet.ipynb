```python
plt.scatter(x=df['year'], y=df['orbital_period'], marker='*')
plt.grid(True)
plt.xlabel('Year')
plt.ylabel('orbital_period')
plt.title('orbital_period by year')
plt.show()
```

```python
x = df['method']
y = df['orbital_period']

plt.bar(x,y, color='red',width=0.5, align='center', alpha=0.5, label='Orbital Period', edgecolor='black')
plt.xlabel('method')
plt.ylabel('orbital_period')
plt.xticks(rotation=90)
plt.title('Orbital Period')
plt.legend()
plt.show()
```

```python
sns.barplot(x='method', y='distance',data=df, palette='summer')
plt.xlabel('method')
plt.ylabel('distance')
plt.xticks(rotation=90)
plt.legend(loc='upper left')
plt.title('bar plot')
plt.show()
```
